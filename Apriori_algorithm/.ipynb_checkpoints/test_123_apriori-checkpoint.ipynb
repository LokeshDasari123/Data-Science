```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
data = pd.read_csv('Groceries_dataset.csv')
```

```python
data
```

```python
data.info()
```

```python
data.describe()
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
sorted_df = df.sort_values(by="Member_number")
sorted_df
```

```python
sorted_df.info()
```

```python
item_list = sorted_df.groupby(["Member_number","Date"])["itemDescription"].agg(lambda x : ','.join(x)).reset_index()
```

```python
item_list
```

```python
item_list.drop(["Member_number","Date"], axis=1,inplace =True)
item_list.columns = ['itemList']
```

```python
item_df = pd.DataFrame(item_list)
item_df
```

```python
from collections import Counter

all_items = [item for sublist in item_df['itemList'].str.split(',') for item in sublist]

# Use Counter to count the frequency of each item
item_frequencies = Counter(all_items)

# Create a DataFrame from the Counter
item_frequency_df = pd.DataFrame(list(item_frequencies.items()), columns=['Item', 'Frequency'])

# Display the frequency of each item
print("Item Frequencies:")
item_frequency_df
```

```python
transactions = item_frequency_df[['Item']].values.tolist()
transactions
```

```python
rules = apriori(transactions,min_support=0.001,min_confidance=0.2,min_lift=1,min_length=2)
rules
```

```python
res = list(rules)
res
```

```python
df_results = pd.DataFrame(res)
df_results
```

```python
df_results.describe()
```

```python
support = df_results.support
```

```python
first_values = []
second_values = []
third_values = []
fourth_value = []

# loop number of rows time and append 1 by 1 value in a separate list.. first and second element was frozenset which need to be converted in list..
for i in range(df_results.shape[0]):
    single_list = df_results['ordered_statistics'][i][0]
    first_values.append(list(single_list[0]))
    second_values.append(list(single_list[1]))
    third_values.append(single_list[2])
    fourth_value.append(single_list[3])
```

```python
lhs = pd.DataFrame(first_values)
rhs= pd.DataFrame(second_values)
confidance=pd.DataFrame(third_values,columns=['Confidance'])
lift=pd.DataFrame(fourth_value,columns=['lift'])
```

```python
df_final = pd.concat([lhs,rhs,support,confidance,lift], axis=1)
df_final
```
