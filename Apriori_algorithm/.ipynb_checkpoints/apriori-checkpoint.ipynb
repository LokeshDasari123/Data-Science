```python
sorted_df = df.sort_values(by="Member_number")
sorted_df
```

```python
sorted_df.info()
```

```python
item_list = sorted_df.groupby(["Member_number","Date"])["itemDescription"].agg(lambda x : ','.join(x)).reset_index()
```

```python
item_list
```

```python
item_list.drop(["Member_number","Date"], axis=1,inplace =True)
item_list.columns = ['itemList']
```

```python
item_df = pd.DataFrame(item_list)
item_df
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
transactions = [item_list.split(',') for item_list in item_df['itemList']]
```

```python
transactions = [list(set(transaction)) for transaction in transactions]
transactions
```

```python
te = TransactionEncoder()
te
```

```python
one_hot = te.fit_transform(transactions)
one_hot
```

```python
txn = pd.DataFrame(one_hot,columns=te.columns_)
txn
```

```python
freq_itemsets = apriori(txn, min_support=0.001, use_colnames=True)
freq_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(freq_itemsets, metric="confidence", min_threshold=0.05)
```
